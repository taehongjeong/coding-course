### 1.2 Hash Table êµ¬í˜„

````python
# ê°œë… 2: Hash Table í™œìš©
def hash_table_examples():
    """
    Pythonì˜ dict, set, defaultdict, Counter í™œìš© ì˜ˆì œ
    
    ì‹œê°„ë³µì¡ë„: O(1) for insertion/deletion/search (average)
    ê³µê°„ë³µì¡ë„: O(n)
    """
    data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
    print(f"ë°ì´í„°: {data}")
    
    # 1. ê¸°ë³¸ ë”•ì…”ë„ˆë¦¬ë¡œ ë¹ˆë„ìˆ˜ ê³„ì‚°
    freq_dict = {}
    for num in data:
        freq_dict[num] = freq_dict.get(num, 0) + 1
    print(f"\n1. dict ì‚¬ìš©: {freq_dict}")
    
    # 2. defaultdict ì‚¬ìš©
    freq_defaultdict = defaultdict(int)
    for num in data:
        freq_defaultdict[num] += 1
    print(f"2. defaultdict ì‚¬ìš©: {dict(freq_defaultdict)}")
    
    # 3. Counter ì‚¬ìš©
    freq_counter = Counter(data)
    print(f"3. Counter ì‚¬ìš©: {freq_counter}")
    print(f"   ê°€ì¥ ë¹ˆë²ˆí•œ 2ê°œ: {freq_counter.most_common(2)}")
    
    # 4. set í™œìš©
    unique_nums = set(data)
    print(f"\n4. setìœ¼ë¡œ ì¤‘ë³µ ì œê±°: {unique_nums}")
    
    # 5. set ì—°ì‚°
    set1 = {1, 2, 3, 4}
    set2 = {3, 4, 5, 6}
    print(f"\n5. set ì—°ì‚°:")
    print(f"   êµì§‘í•©: {set1 & set2}")
    print(f"   í•©ì§‘í•©: {set1 | set2}")
    print(f"   ì°¨ì§‘í•©: {set1 - set2}")

# í…ŒìŠ¤íŠ¸
print("=== Hash Table í™œìš© ì˜ˆì œ ===")
hash_table_examples()
````

### 2.1 íŒ¨í„´ 1: Two Pointer

````python
# íŒ¨í„´ 1: Two Pointer í…œí”Œë¦¿
def two_pointer_examples():
    """
    Two Pointer ê¸°ë²•ì˜ ë‹¤ì–‘í•œ í™œìš© ì˜ˆì œ
    
    ì‹œê°„ë³µì¡ë„: O(n)
    ê³µê°„ë³µì¡ë„: O(1)
    """
    
    # ì˜ˆì œ 1: ì •ë ¬ëœ ë°°ì—´ì—ì„œ ë‘ ìˆ˜ì˜ í•© ì°¾ê¸°
    def two_sum_sorted(nums: List[int], target: int) -> List[int]:
        left, right = 0, len(nums) - 1
        
        while left < right:
            current_sum = nums[left] + nums[right]
            if current_sum == target:
                return [left, right]
            elif current_sum < target:
                left += 1
            else:
                right -= 1
        return []
    
    # ì˜ˆì œ 2: íŒ°ë¦°ë“œë¡¬ í™•ì¸
    def is_palindrome(s: str) -> bool:
        # ì•ŒíŒŒë²³ê³¼ ìˆ«ìë§Œ ë‚¨ê¸°ê³  ì†Œë¬¸ìë¡œ ë³€í™˜
        s = ''.join(c.lower() for c in s if c.isalnum())
        left, right = 0, len(s) - 1
        
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True
    
    # ì˜ˆì œ 3: Container With Most Water (LeetCode 11)
    def max_area(height: List[int]) -> int:
        left, right = 0, len(height) - 1
        max_water = 0
        
        while left < right:
            width = right - left
            min_height = min(height[left], height[right])
            max_water = max(max_water, width * min_height)
            
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        
        return max_water
    
    # í…ŒìŠ¤íŠ¸
    print("1. Two Sum (ì •ë ¬ëœ ë°°ì—´):")
    nums = [1, 2, 7, 11, 15]
    target = 9
    result = two_sum_sorted(nums, target)
    print(f"   ì…ë ¥: {nums}, target={target}")
    print(f"   ê²°ê³¼: ì¸ë±ìŠ¤ {result} -> [{nums[result[0]]}, {nums[result[1]]}]\n")
    
    print("2. íŒ°ë¦°ë“œë¡¬ í™•ì¸:")
    test_strings = ["A man, a plan, a canal: Panama", "race a car"]
    for s in test_strings:
        print(f"   \"{s}\" -> {is_palindrome(s)}")
    
    print("\n3. Container With Most Water:")
    height = [1, 8, 6, 2, 5, 4, 8, 3, 7]
    print(f"   ë†’ì´: {height}")
    print(f"   ìµœëŒ€ ë¬¼ì˜ ì–‘: {max_area(height)}")

# íŒ¨í„´ ì ìš© ì˜ˆì‹œ
print("=== Two Pointer íŒ¨í„´ ì˜ˆì‹œ ===")
two_pointer_examples()
````

### 2.2 íŒ¨í„´ 2: Sliding Window

````python
# íŒ¨í„´ 2: Sliding Window í…œí”Œë¦¿
def sliding_window_examples():
    """
    Sliding Window ê¸°ë²•ì˜ ë‹¤ì–‘í•œ í™œìš© ì˜ˆì œ
    
    ì‹œê°„ë³µì¡ë„: O(n)
    ê³µê°„ë³µì¡ë„: O(1) or O(k)
    """
    
    # ì˜ˆì œ 1: ê³ ì • í¬ê¸° ìœˆë„ìš° - ìµœëŒ€ í•©
    def max_sum_subarray(nums: List[int], k: int) -> int:
        if len(nums) < k:
            return 0
        
        # ì´ˆê¸° ìœˆë„ìš°
        window_sum = sum(nums[:k])
        max_sum = window_sum
        
        # ìŠ¬ë¼ì´ë”©
        for i in range(k, len(nums)):
            window_sum = window_sum - nums[i-k] + nums[i]
            max_sum = max(max_sum, window_sum)
        
        return max_sum
    
    # ì˜ˆì œ 2: ê°€ë³€ í¬ê¸° ìœˆë„ìš° - ìµœì¥ ë¶€ë¶„ ë¬¸ìì—´ (ì¤‘ë³µ ì—†ìŒ)
    def longest_substring_without_repeating(s: str) -> int:
        char_set = set()
        left = 0
        max_length = 0
        
        for right in range(len(s)):
            # ì¤‘ë³µ ë¬¸ìê°€ ìˆìœ¼ë©´ ì™¼ìª½ í¬ì¸í„° ì´ë™
            while s[right] in char_set:
                char_set.remove(s[left])
                left += 1
            
            char_set.add(s[right])
            max_length = max(max_length, right - left + 1)
        
        return max_length
    
    # ì˜ˆì œ 3: ê°€ë³€ í¬ê¸° ìœˆë„ìš° - ìµœì†Œ ê¸¸ì´ ë¶€ë¶„ ë°°ì—´
    def min_subarray_len(target: int, nums: List[int]) -> int:
        left = 0
        current_sum = 0
        min_length = float('inf')
        
        for right in range(len(nums)):
            current_sum += nums[right]
            
            while current_sum >= target:
                min_length = min(min_length, right - left + 1)
                current_sum -= nums[left]
                left += 1
        
        return min_length if min_length != float('inf') else 0
    
    # í…ŒìŠ¤íŠ¸
    print("1. ê³ ì • í¬ê¸° ìœˆë„ìš° - ìµœëŒ€ í•©:")
    nums = [2, 1, 5, 1, 3, 2]
    k = 3
    print(f"   ë°°ì—´: {nums}, k={k}")
    print(f"   í¬ê¸° {k}ì¸ ë¶€ë¶„ ë°°ì—´ì˜ ìµœëŒ€ í•©: {max_sum_subarray(nums, k)}\n")
    
    print("2. ê°€ë³€ í¬ê¸° ìœˆë„ìš° - ì¤‘ë³µ ì—†ëŠ” ìµœì¥ ë¶€ë¶„ ë¬¸ìì—´:")
    test_strings = ["abcabcbb", "bbbbb", "pwwkew"]
    for s in test_strings:
        print(f"   \"{s}\" -> ê¸¸ì´ {longest_substring_without_repeating(s)}")
    
    print("\n3. ê°€ë³€ í¬ê¸° ìœˆë„ìš° - í•©ì´ target ì´ìƒì¸ ìµœì†Œ ê¸¸ì´:")
    nums = [2, 3, 1, 2, 4, 3]
    target = 7
    print(f"   ë°°ì—´: {nums}, target={target}")
    print(f"   ìµœì†Œ ê¸¸ì´: {min_subarray_len(target, nums)}")

print("=== Sliding Window íŒ¨í„´ ì˜ˆì‹œ ===")
sliding_window_examples()
````

### Problem 1: [1] Two Sum

````python
"""
ë¬¸ì œ ì„¤ëª…:
ì •ìˆ˜ ë°°ì—´ numsì™€ ì •ìˆ˜ targetì´ ì£¼ì–´ì§ˆ ë•Œ, ë‘ ìˆ˜ì˜ í•©ì´ targetì´ ë˜ëŠ” 
ë‘ ìˆ˜ì˜ ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜í•˜ë¼.

ì˜ˆì‹œ:
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
ì„¤ëª…: nums[0] + nums[1] = 2 + 7 = 9

ì œì•½ì‚¬í•­:
- 2 <= nums.length <= 10^4
- -10^9 <= nums[i] <= 10^9
- ì •ë‹µì€ ë‹¨ í•˜ë‚˜ë§Œ ì¡´ì¬
"""

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        """
        ì ‘ê·¼ë²•:
        1. Hash Tableì„ ì‚¬ìš©í•˜ì—¬ ì´ë¯¸ ë³¸ ìˆ«ìì™€ ì¸ë±ìŠ¤ ì €ì¥
        2. ê° ìˆ«ìì— ëŒ€í•´ complement(target - num) í™•ì¸
        3. complementê°€ Hash Tableì— ìˆìœ¼ë©´ ì¸ë±ìŠ¤ ë°˜í™˜
        
        ì‹œê°„ë³µì¡ë„: O(n)
        ê³µê°„ë³µì¡ë„: O(n)
        """
        seen = {}  # ê°’: ì¸ë±ìŠ¤
        
        for i, num in enumerate(nums):
            complement = target - num
            if complement in seen:
                return [seen[complement], i]
            seen[num] = i
        
        return []  # í•´ë‹µì´ ì—†ëŠ” ê²½ìš° (ë¬¸ì œ ì¡°ê±´ìƒ ë°œìƒí•˜ì§€ ì•ŠìŒ)

# í…ŒìŠ¤íŠ¸
solution = Solution()
test_cases = [
    ([2, 7, 11, 15], 9, [0, 1]),
    ([3, 2, 4], 6, [1, 2]),
    ([3, 3], 6, [0, 1])
]

print("=== Two Sum í…ŒìŠ¤íŠ¸ ===")
for i, (nums, target, expected) in enumerate(test_cases, 1):
    result = solution.twoSum(nums, target)
    print(f"í…ŒìŠ¤íŠ¸ {i}: {'âœ…' if result == expected else 'âŒ'}")
    print(f"  ì…ë ¥: nums={nums}, target={target}")
    print(f"  ì¶œë ¥: {result}")
    print(f"  ì˜ˆìƒ: {expected}\n")
````

### Problem 2: [242] Valid Anagram

````python
"""
ë¬¸ì œ ì„¤ëª…:
ë‘ ë¬¸ìì—´ sì™€ tê°€ ì£¼ì–´ì§ˆ ë•Œ, tê°€ sì˜ anagramì¸ì§€ í™•ì¸í•˜ë¼.
Anagram: ê°™ì€ ë¬¸ìë“¤ì„ ì¬ë°°ì—´í•˜ì—¬ ë§Œë“  ë‹¨ì–´

ì˜ˆì‹œ:
Input: s = "anagram", t = "nagaram"
Output: true

ì œì•½ì‚¬í•­:
- 1 <= s.length, t.length <= 5 * 10^4
- sì™€ tëŠ” ì†Œë¬¸ì ì˜ì–´ë¡œë§Œ êµ¬ì„±
"""

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        """
        ì ‘ê·¼ë²• 1: Counter ì‚¬ìš©
        ë‘ ë¬¸ìì—´ì˜ ë¬¸ì ë¹ˆë„ìˆ˜ê°€ ê°™ì€ì§€ í™•ì¸
        
        ì‹œê°„ë³µì¡ë„: O(n)
        ê³µê°„ë³µì¡ë„: O(1) - ìµœëŒ€ 26ê°œì˜ ë¬¸ìë§Œ ì €ì¥
        """
        # ë°©ë²• 1: Counter ì‚¬ìš©
        return Counter(s) == Counter(t)
    
    def isAnagram_v2(self, s: str, t: str) -> bool:
        """
        ì ‘ê·¼ë²• 2: ì •ë ¬ ì‚¬ìš©
        
        ì‹œê°„ë³µì¡ë„: O(n log n)
        ê³µê°„ë³µì¡ë„: O(1)
        """
        return sorted(s) == sorted(t)
    
    def isAnagram_v3(self, s: str, t: str) -> bool:
        """
        ì ‘ê·¼ë²• 3: ìˆ˜ë™ìœ¼ë¡œ ë¹ˆë„ìˆ˜ ê³„ì‚°
        
        ì‹œê°„ë³µì¡ë„: O(n)
        ê³µê°„ë³µì¡ë„: O(1)
        """
        if len(s) != len(t):
            return False
        
        count = {}
        for char in s:
            count[char] = count.get(char, 0) + 1
        
        for char in t:
            if char not in count:
                return False
            count[char] -= 1
            if count[char] < 0:
                return False
        
        return True

# í…ŒìŠ¤íŠ¸
solution = Solution()
test_cases = [
    ("anagram", "nagaram", True),
    ("rat", "car", False),
    ("listen", "silent", True)
]

print("=== Valid Anagram í…ŒìŠ¤íŠ¸ ===")
for i, (s, t, expected) in enumerate(test_cases, 1):
    result = solution.isAnagram(s, t)
    print(f"í…ŒìŠ¤íŠ¸ {i}: {'âœ…' if result == expected else 'âŒ'}")
    print(f"  s = \"{s}\", t = \"{t}\"")
    print(f"  ê²°ê³¼: {result} (ì˜ˆìƒ: {expected})\n")
````

### 1.1 Queue (í) ê¸°ë³¸ êµ¬í˜„

````python
# Queue: FIFO ìë£Œêµ¬ì¡° êµ¬í˜„
class Queue:
    """
    Queue êµ¬í˜„:
    - FIFO (First In First Out) ì›ì¹™
    - deque ì‚¬ìš©ìœ¼ë¡œ O(1) ì—°ì‚° ë³´ì¥
    
    ì‹œê°„ë³µì¡ë„: ëª¨ë“  ì—°ì‚° O(1)
    ê³µê°„ë³µì¡ë„: O(n)
    """
    def __init__(self):
        self.items = deque()
    
    def enqueue(self, item):
        """ì›ì†Œ ì¶”ê°€"""
        self.items.append(item)
    
    def dequeue(self):
        """ì›ì†Œ ì œê±°"""
        if not self.is_empty():
            return self.items.popleft()
        return None
    
    def peek(self):
        """ì²« ë²ˆì§¸ ì›ì†Œ í™•ì¸"""
        if not self.is_empty():
            return self.items[0]
        return None
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
    
    def __str__(self):
        return str(list(self.items))

# í…ŒìŠ¤íŠ¸
print("=== Queue í…ŒìŠ¤íŠ¸ ===")
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(f"í ìƒíƒœ: {q}")
print(f"dequeue: {q.dequeue()}")
print(f"peek: {q.peek()}")
print(f"í ìƒíƒœ: {q}")
print(f"í¬ê¸°: {q.size()}")
````

### 1.2 Priority Queue (ìš°ì„ ìˆœìœ„ í) êµ¬í˜„

````python
# Priority Queue: heapqë¥¼ ì´ìš©í•œ êµ¬í˜„
class PriorityQueue:
    """
    ìš°ì„ ìˆœìœ„ í êµ¬í˜„:
    - ìµœì†Œ í™ ê¸°ë°˜
    - ìš°ì„ ìˆœìœ„ê°€ ë‚®ì„ìˆ˜ë¡ ë¨¼ì € ë‚˜ì˜´
    
    ì‹œê°„ë³µì¡ë„: push/pop O(log n)
    ê³µê°„ë³µì¡ë„: O(n)
    """
    def __init__(self):
        self.heap = []
        self.index = 0  # ë™ì¼ ìš°ì„ ìˆœìœ„ ì²˜ë¦¬ìš©
    
    def push(self, item, priority):
        """ìš°ì„ ìˆœìœ„ì™€ í•¨ê»˜ ì›ì†Œ ì¶”ê°€"""
        heappush(self.heap, (priority, self.index, item))
        self.index += 1
    
    def pop(self):
        """ê°€ì¥ ë†’ì€ ìš°ì„ ìˆœìœ„ ì›ì†Œ ì œê±°"""
        if self.heap:
            return heappop(self.heap)[2]
        return None
    
    def peek(self):
        """ê°€ì¥ ë†’ì€ ìš°ì„ ìˆœìœ„ ì›ì†Œ í™•ì¸"""
        if self.heap:
            return self.heap[0][2]
        return None
    
    def is_empty(self):
        return len(self.heap) == 0
    
    def size(self):
        return len(self.heap)

# í…ŒìŠ¤íŠ¸
print("=== Priority Queue í…ŒìŠ¤íŠ¸ ===")
pq = PriorityQueue()
pq.push("ì‘ì—…3", 3)
pq.push("ì‘ì—…1", 1)
pq.push("ì‘ì—…2", 2)
pq.push("ê¸´ê¸‰ì‘ì—…", 1)  # ë™ì¼ ìš°ì„ ìˆœìœ„

print("ìš°ì„ ìˆœìœ„ ìˆœì„œëŒ€ë¡œ ì²˜ë¦¬:")
while not pq.is_empty():
    print(f"  {pq.pop()}")
````

### 2.1 íŒ¨í„´ 1: Sorting Algorithms ë¹„êµ

````python
# ì •ë ¬ ì•Œê³ ë¦¬ì¦˜ êµ¬í˜„ ë° ì„±ëŠ¥ ë¹„êµ
import random
import time

def bubble_sort(arr):
    """
    ë²„ë¸” ì •ë ¬:
    - ì¸ì ‘í•œ ì›ì†Œ ë¹„êµ ë° êµí™˜
    - ì•ˆì • ì •ë ¬
    
    ì‹œê°„ë³µì¡ë„: O(nÂ²)
    ê³µê°„ë³µì¡ë„: O(1)
    """
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

def quick_sort(arr):
    """
    í€µ ì •ë ¬:
    - ë¶„í•  ì •ë³µ ë°©ì‹
    - ë¶ˆì•ˆì • ì •ë ¬
    
    ì‹œê°„ë³µì¡ë„: í‰ê·  O(n log n), ìµœì•… O(nÂ²)
    ê³µê°„ë³µì¡ë„: O(log n)
    """
    if len(arr) <= 1:
        return arr
    
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    return quick_sort(left) + middle + quick_sort(right)

def merge_sort(arr):
    """
    ë³‘í•© ì •ë ¬:
    - ë¶„í•  ì •ë³µ ë°©ì‹
    - ì•ˆì • ì •ë ¬
    
    ì‹œê°„ë³µì¡ë„: O(n log n)
    ê³µê°„ë³µì¡ë„: O(n)
    """
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    # ë³‘í•©
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# ì„±ëŠ¥ ë¹„êµ
print("=== ì •ë ¬ ì•Œê³ ë¦¬ì¦˜ ì„±ëŠ¥ ë¹„êµ ===")

# ì‘ì€ ë°°ì—´ í…ŒìŠ¤íŠ¸
small_arr = [64, 34, 25, 12, 22, 11, 90]
print(f"ì›ë³¸ ë°°ì—´: {small_arr}")
print(f"ë²„ë¸” ì •ë ¬: {bubble_sort(small_arr)}")
print(f"í€µ ì •ë ¬: {quick_sort(small_arr)}")
print(f"ë³‘í•© ì •ë ¬: {merge_sort(small_arr)}")
print(f"íŒŒì´ì¬ ì •ë ¬: {sorted(small_arr)}")

# ì„±ëŠ¥ ì¸¡ì •
print("\n=== ì„±ëŠ¥ ì¸¡ì • (n=100) ===")
test_arr = [random.randint(1, 1000) for _ in range(100)]

# ë²„ë¸” ì •ë ¬ (ì‘ì€ ë°ì´í„°ì—ë§Œ)
start = time.time()
bubble_sort(test_arr.copy())
print(f"ë²„ë¸” ì •ë ¬: {(time.time() - start)*1000:.2f}ms")

# í€µ ì •ë ¬
start = time.time()
quick_sort(test_arr.copy())
print(f"í€µ ì •ë ¬: {(time.time() - start)*1000:.2f}ms")

# ë³‘í•© ì •ë ¬
start = time.time()
merge_sort(test_arr.copy())
print(f"ë³‘í•© ì •ë ¬: {(time.time() - start)*1000:.2f}ms")

# íŒŒì´ì¬ ë‚´ì¥ ì •ë ¬
start = time.time()
sorted(test_arr.copy())
print(f"íŒŒì´ì¬ ì •ë ¬: {(time.time() - start)*1000:.2f}ms")
````

### 2.2 íŒ¨í„´ 2: BFS with Queue

````python
# BFSë¥¼ ì´ìš©í•œ ë ˆë²¨ ìˆœíšŒì™€ ìµœë‹¨ ê²½ë¡œ
from collections import deque

def bfs_level_order(graph, start):
    """
    BFS ë ˆë²¨ë³„ ìˆœíšŒ:
    - íë¥¼ ì‚¬ìš©í•œ ë„ˆë¹„ ìš°ì„  íƒìƒ‰
    - ê° ë ˆë²¨ë³„ë¡œ ë…¸ë“œ êµ¬ë¶„
    
    ì‹œê°„ë³µì¡ë„: O(V + E)
    ê³µê°„ë³µì¡ë„: O(V)
    """
    if start not in graph:
        return []
    
    visited = set([start])
    queue = deque([start])
    levels = []
    
    while queue:
        level_size = len(queue)
        current_level = []
        
        for _ in range(level_size):
            node = queue.popleft()
            current_level.append(node)
            
            for neighbor in graph[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)
        
        levels.append(current_level)
    
    return levels

def shortest_path_bfs(graph, start, end):
    """
    BFSë¥¼ ì´ìš©í•œ ìµœë‹¨ ê²½ë¡œ ì°¾ê¸°:
    - ë¬´ê°€ì¤‘ ê·¸ë˜í”„ì—ì„œ ìµœë‹¨ ê²½ë¡œ ë³´ì¥
    - ê²½ë¡œì™€ ê±°ë¦¬ ë°˜í™˜
    
    ì‹œê°„ë³µì¡ë„: O(V + E)
    ê³µê°„ë³µì¡ë„: O(V)
    """
    if start == end:
        return [start], 0
    
    visited = set([start])
    queue = deque([(start, [start])])
    
    while queue:
        node, path = queue.popleft()
        
        for neighbor in graph.get(node, []):
            if neighbor == end:
                return path + [end], len(path)
            
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    
    return None, -1  # ê²½ë¡œ ì—†ìŒ

# í…ŒìŠ¤íŠ¸
print("=== BFS íŒ¨í„´ ì˜ˆì‹œ ===")

# ê·¸ë˜í”„ ì •ì˜ (ì¸ì ‘ ë¦¬ìŠ¤íŠ¸)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# ë ˆë²¨ë³„ ìˆœíšŒ
levels = bfs_level_order(graph, 'A')
print("ë ˆë²¨ë³„ BFS ìˆœíšŒ:")
for i, level in enumerate(levels):
    print(f"  Level {i}: {level}")

# ìµœë‹¨ ê²½ë¡œ ì°¾ê¸°
print("\nìµœë‹¨ ê²½ë¡œ ì°¾ê¸°:")
test_cases = [('A', 'F'), ('A', 'D'), ('B', 'C')]
for start, end in test_cases:
    path, distance = shortest_path_bfs(graph, start, end)
    if path:
        print(f"  {start} â†’ {end}: ê²½ë¡œ={' â†’ '.join(path)}, ê±°ë¦¬={distance}")
    else:
        print(f"  {start} â†’ {end}: ê²½ë¡œ ì—†ìŒ")
````

### Problem 1: [56] Merge Intervals

````python
"""
ë¬¸ì œ ì„¤ëª…:
- ê²¹ì¹˜ëŠ” êµ¬ê°„ë“¤ì„ ë³‘í•©í•˜ì—¬ ë°˜í™˜

ì˜ˆì‹œ:
Input: intervals = [[1,3],[2,6],[8,10],[15,18]]
Output: [[1,6],[8,10],[15,18]]

ì œì•½ì‚¬í•­:
- 1 <= intervals.length <= 10^4
- intervals[i].length == 2
"""

class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        """
        ì ‘ê·¼ë²•:
        1. ì‹œì‘ì  ê¸°ì¤€ìœ¼ë¡œ ì •ë ¬
        2. ìˆœì°¨ì ìœ¼ë¡œ êµ¬ê°„ í™•ì¸í•˜ë©° ë³‘í•©
        3. ê²¹ì¹˜ë©´ ë³‘í•©, ì•ˆ ê²¹ì¹˜ë©´ ìƒˆë¡œ ì¶”ê°€
        
        ì‹œê°„ë³µì¡ë„: O(n log n)
        ê³µê°„ë³µì¡ë„: O(n)
        """
        if not intervals:
            return []
        
        # ì‹œì‘ì  ê¸°ì¤€ ì •ë ¬
        intervals.sort(key=lambda x: x[0])
        
        merged = [intervals[0]]
        
        for current in intervals[1:]:
            last = merged[-1]
            
            # ê²¹ì¹˜ëŠ” ê²½ìš° ë³‘í•©
            if current[0] <= last[1]:
                merged[-1] = [last[0], max(last[1], current[1])]
            else:
                # ê²¹ì¹˜ì§€ ì•Šìœ¼ë©´ ì¶”ê°€
                merged.append(current)
        
        return merged

# í…ŒìŠ¤íŠ¸
solution = Solution()
test_cases = [
    ([[1,3],[2,6],[8,10],[15,18]], [[1,6],[8,10],[15,18]]),
    ([[1,4],[4,5]], [[1,5]]),
    ([[1,4],[0,4]], [[0,4]]),
    ([[1,4],[2,3]], [[1,4]])
]

for i, (input_data, expected) in enumerate(test_cases, 1):
    result = solution.merge(input_data)
    print(f"í…ŒìŠ¤íŠ¸ {i}: {'âœ…' if result == expected else 'âŒ'}")
    print(f"  ì…ë ¥: {input_data}")
    print(f"  ì¶œë ¥: {result}")
    print(f"  ì˜ˆìƒ: {expected}\n")
````

### Problem 2: [215] Kth Largest Element in an Array

````python
"""
ë¬¸ì œ ì„¤ëª…:
- ë°°ì—´ì—ì„œ kë²ˆì§¸ë¡œ í° ì›ì†Œë¥¼ ì°¾ì•„ ë°˜í™˜

ì˜ˆì‹œ:
Input: nums = [3,2,1,5,6,4], k = 2
Output: 5

ì œì•½ì‚¬í•­:
- 1 <= k <= nums.length <= 10^4
- -10^4 <= nums[i] <= 10^4
"""

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        """
        ì ‘ê·¼ë²• 1: í™ ì‚¬ìš© (ìµœì†Œ í™)
        1. í¬ê¸° kì˜ ìµœì†Œ í™ ìœ ì§€
        2. í™ì˜ ë£¨íŠ¸ê°€ kë²ˆì§¸ í° ì›ì†Œ
        
        ì‹œê°„ë³µì¡ë„: O(n log k)
        ê³µê°„ë³µì¡ë„: O(k)
        """
        import heapq
        
        # ìµœì†Œ í™ ì‚¬ìš©
        heap = []
        
        for num in nums:
            heapq.heappush(heap, num)
            if len(heap) > k:
                heapq.heappop(heap)
        
        return heap[0]
    
    def findKthLargestQuickSelect(self, nums: List[int], k: int) -> int:
        """
        ì ‘ê·¼ë²• 2: Quick Select
        í‰ê·  ì‹œê°„ë³µì¡ë„: O(n)
        ìµœì•… ì‹œê°„ë³µì¡ë„: O(nÂ²)
        """
        def quick_select(left, right, k_smallest):
            if left == right:
                return nums[left]
            
            # í”¼ë²— ì„ íƒ ë° íŒŒí‹°ì…˜
            pivot_index = partition(left, right)
            
            if k_smallest == pivot_index:
                return nums[k_smallest]
            elif k_smallest < pivot_index:
                return quick_select(left, pivot_index - 1, k_smallest)
            else:
                return quick_select(pivot_index + 1, right, k_smallest)
        
        def partition(left, right):
            pivot = nums[right]
            store_index = left
            
            for i in range(left, right):
                if nums[i] < pivot:
                    nums[store_index], nums[i] = nums[i], nums[store_index]
                    store_index += 1
            
            nums[right], nums[store_index] = nums[store_index], nums[right]
            return store_index
        
        return quick_select(0, len(nums) - 1, len(nums) - k)

# í…ŒìŠ¤íŠ¸
solution = Solution()
test_cases = [
    ([3,2,1,5,6,4], 2, 5),
    ([3,2,3,1,2,4,5,5,6], 4, 4),
    ([1], 1, 1),
    ([2,1], 2, 1)
]

print("=== Heap ë°©ë²• ===")
for i, (nums, k, expected) in enumerate(test_cases, 1):
    result = solution.findKthLargest(nums.copy(), k)
    print(f"í…ŒìŠ¤íŠ¸ {i}: {'âœ…' if result == expected else 'âŒ'}")
    print(f"  ì…ë ¥: nums={nums}, k={k}")
    print(f"  ì¶œë ¥: {result}")
    print(f"  ì˜ˆìƒ: {expected}\n")
````
